```python
from bs4 import BeautifulSoup 
import requests 
import pandas as pd 
```

```python
# URL of the IMDB list
url = "https://www.imdb.com/list/ls056092300/"

# Headers to mimic a browser visit
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}

# Sending a request to fetch the webpage content
response = requests.get(url, headers=headers)

 # Check if the request was successful
if response.status_code == 200:
    print("Success")
else:
    print("Connection fialed with status code: {response.status_code}")
```

```python
if response.status_code == 200:
    # Parsing the content with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
# Finding the movie name containers
movie_name_containers = soup.find_all('div', class_="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-b189961a-9 iALATN dli-title")
movie_name_containers
```

```python
#Fetch the movie raings
ratings_containers = soup.find_all('div', class_="sc-e2dbc1a3-0 ajrIH sc-b189961a-2 fkPBP dli-ratings-container")
ratings_containers
```

```python
#Fetch the movies years
years_containers = soup.find_all('div', class_="sc-b189961a-7 feoqjK dli-title-metadata")
years_containers
```

```python
#Create empty lsit to hold items 
movieNames = []
moviesRatings = []
yearOfRelease = []

    # Extracting the required data and appending to list
for container in movie_name_containers:
        # Movie name
        name_tag = container.find('h3', class_='ipc-title__text')
        name = name_tag.text.strip() if name_tag else 'N/A'
        movieNames.append(name)
```

```python
movieNames
```

```python
  # Movie rating
for container in ratings_containers:
   # Movie rating
        rating_tag = container.find('span', {'aria-label': lambda x: x and x.startswith('IMDb rating:')})
        rating = rating_tag['aria-label'].split(':')[-1].strip() if rating_tag else 'N/A'
        moviesRatings.append(rating)
```

```python
moviesRatings
```

```python

    # Extracting the required data and appending to list
for container in years_containers:
        # year of release
        year_tag = container.find('span', class_='sc-b189961a-8 kLaxqf dli-title-metadata-item')
        year = year_tag.text.strip() if year_tag else 'N/A'
        yearOfRelease.append(year)
```

```python
yearOfRelease
```

```python
   # Creating a DataFrame
movies_df = pd.DataFrame({
        'Name': movieNames,
        'Rating': moviesRatings,
        'Year of Release': yearOfRelease
    })


```

```python
!pip install selenium

from selenium import webdriver
from selenium.webdriver.firefox.service import Service as FirefoxService
from selenium.webdriver.firefox.options import Options
import time
```

```python
# Setup Firefox options
firefox_options = Options()
firefox_options.add_argument("--headless")  # Ensure GUI is off

# Set path to geckodriver as per your configuration
webdriver_service = FirefoxService(executable_path="/Users/mac/Downloads/geckodriver")

# Define the URL
task2_url = "https://www.patreon.com/coreyms"


# Send a request to check if the page is accessible
response = requests.get(task2_url)

# Check if the request was successful
if response.status_code == 200:
    print("Success")
else:
    print("Failed connection with status code: {response.status_code}")
```

```python

# Using a context manager to ensure the driver is managed and closed properly
with webdriver.Firefox(service=webdriver_service, options=firefox_options) as driver:
    # Open the webpage
    driver.get(task2_url)

    # Allow some time for the page to load
    time.sleep(5)  # You may need to increase this time depending on your internet speed

    # Scroll to the bottom of the page to load all posts (you may need to adjust the range)
    for _ in range(10):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)

    # Get page source and create BeautifulSoup object
    soup_2 = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
soup_2
```

```python
# No need to explicitly call driver.quit(), the context manager handles it

# Check if the driver has been quit
try:
    driver.current_url
    print("Driver has not been quit.")
except (AttributeError, Exception):
    print("Driver has been quit.")
```

```python
# Continue processing the scraped data outside the context manager

  # Find all posts
posts = soup_2.select('div.sc-bBHxTw.iloeMK')
print(f"Found {len(posts)} posts")
```

```python
# Find all posts
allPosts = soup_2.find_all('div', class_='sc-bBHxTw iloeMK')
allPosts
```

```python
# create empty Lists to store data

allHeadings = []
allDates = []
allContents = []
allLikes = []
```

```python
#Retrieve heading and append to list

for headings in allPosts:
        heading_tag = headings.find('span', {'data-tag': 'post-title'})
        heading = heading_tag.text.strip() if heading_tag else 'N/A'
        allHeadings.append(heading)
```

```python
allHeadings
```

```python
# Find all dates of posts
dates = soup_2.find_all('div', class_='sc-lgu5zg-0 dXpjXs')
dates
```

```python
#Retrieve dates and append to list

for date in dates:
    date_tag = date.find('a', {'data-tag': 'post-published-at'})
    date_span = date_tag.find('span').find('span') if date_tag else None
    postDate = date_span.text.strip() if date_span else 'N/A'
    allDates.append(postDate)
```

```python
allDates
```

```python
#Fetch the contents of posts
contents = soup_2.find_all('div', class_='sc-cfnzm4-0 daxSFj')
contents
```

```python
#Retrieve content and append to list
for content in contents:
    p_tag = content.find('p')
    p_content = p_tag.text.strip() if p_tag else 'N/A'
    allContents.append(p_content)
```

```python
allContents
```

```python
#Fetch total likes

likes = soup_2.find_all('div', class_='sc-jrQzAO knZGTZ')
likes
```

```python
#Search all the likes per post

allLikes = []

for like in likes:
    like_tag = like.find('span', {'data-tag': 'like-count'})
    like_count = like_tag.text.strip() if like_tag else 'N/A'
    allLikes.append(like_count)
```

```python
allLikes
```

```python
# Creating the Dataframe for Posts
posts_statistics = pd.DataFrame({
        'Headings': allHeadings,
        'Comments': allContents,
        'Date of Post': allDates,
        'Number of likes': allLikes
    })

print('Dataframe created')
```
